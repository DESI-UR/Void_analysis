```python
import pickle

import numpy as np

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

# Data

```python
infile = open('SDSS_DR12r_bluehive_multi_profiling.pickle', 'rb')
num_cells_processed_multi = np.array(pickle.load(infile))
infile.close()

infile = open('SDSS_DR12r_bluehive_multi_90_profiling.pickle', 'rb')
num_cells_processed_multi_90 = np.array(pickle.load(infile))
infile.close()

infile = open('SDSS_DR12r_Steve_multi_thread_profile.pickle', 'rb')
num_cells_processed_multi_Steve = np.array(pickle.load(infile))
infile.close()

infile = open('SDSS_DR12r_bluehive_single_profiling.pickle', 'rb')
num_cells_processed_single = np.array(pickle.load(infile))
infile.close()

infile = open('SDSS_DR12r_Steve_single_thread_profile.pickle', 'rb')
num_cells_processed_single_Steve = np.array(pickle.load(infile))
infile.close()
```

```python
print(len(num_cells_processed_multi))
print(len(num_cells_processed_multi_90))
print(len(num_cells_processed_single))
```

```python
time_multi = np.arange(len(num_cells_processed_multi))*5
time_multi_90 = np.arange(len(num_cells_processed_multi_90))*5
time_multi_Steve = np.arange(len(num_cells_processed_multi_Steve))*5
time_single = np.arange(len(num_cells_processed_single))*5
time_single_Steve = np.arange(len(num_cells_processed_single_Steve))*5
```

# Number of processed cells

```python
plt.figure()
plt.plot(time_multi, num_cells_processed_multi, label='Multi - 80G')
#plt.plot(time_multi_90, num_cells_processed_multi_90, label='Multi - 90G')
plt.plot(time_multi_Steve, num_cells_processed_multi_Steve, label='Multi - Steve')
plt.plot(time_single, num_cells_processed_single, label='Single')
plt.plot(time_single_Steve, num_cells_processed_single_Steve, label='Single - Steve')
plt.xlabel('Time elapsed [s]')
plt.ylabel('Cumulative number of cells processed')
plt.legend()
```

# Rate of cells processed

```python
plt.figure()
plt.plot(time_multi[1:], np.diff(num_cells_processed_multi)/5, label='Multi - 80G')
#plt.plot(time_multi_90[1:], np.diff(num_cells_processed_multi_90)/5, label='Multi - 90G')
plt.plot(time_multi_Steve[1:], np.diff(num_cells_processed_multi_Steve)/5, label='Multi - Steve')
plt.plot(time_single[1:], np.diff(num_cells_processed_single)/5, label='Single')
plt.plot(time_single_Steve[1:], np.diff(num_cells_processed_single_Steve)/5, label='Single - Steve')
plt.xlabel('Time elapsed [s]')
plt.ylabel('Cell processing rate [cells/s]')
plt.legend()
```
