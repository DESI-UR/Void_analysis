```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
#read in file with maximal holes

file_name = 'Data/dr7Data_comoving_maximal_edge_length_'

table_2 = Table.read(file_name+str(2)+'.txt', format='ascii.commented_header')

#table_5 = Table.read(file_name+str(5)+'.txt', format='ascii.commented_header')
#table_6 = Table.read(file_name+str(6)+'.txt', format='ascii.commented_header')
#table_7 = Table.read(file_name+str(7)+'.txt', format='ascii.commented_header')
#table_8 = Table.read(file_name+str(8)+'.txt', format='ascii.commented_header')
#table_9 = Table.read(file_name+str(9)+'.txt', format='ascii.commented_header')
#table_10 = Table.read(file_name+str(10)+'.txt', format='ascii.commented_header')
#table_11 = Table.read(file_name+str(11)+'.txt', format='ascii.commented_header')
#table_12 = Table.read(file_name+str(12)+'.txt', format='ascii.commented_header')
#table_13 = Table.read(file_name+str(13)+'.txt', format='ascii.commented_header')
#table_14 = Table.read(file_name+str(14)+'.txt', format='ascii.commented_header')
#table_15 = Table.read(file_name+str(15)+'.txt', format='ascii.commented_header')
#table_16 = Table.read(file_name+str(16)+'.txt', format='ascii.commented_header')
#table_17 = Table.read(file_name+str(17)+'.txt', format='ascii.commented_header')
#table_18 = Table.read(file_name+str(18)+'.txt', format='ascii.commented_header')
#table_19 = Table.read(file_name+str(19)+'.txt', format='ascii.commented_header')
#table_20 = Table.read(file_name+str(20)+'.txt', format='ascii.commented_header')
```

```python
#build grid size array

number = max(table_2['flag']) + 1

Sizes = np.zeros(19)

for a in range(19):
    Sizes[a] = a+2
print(number)
print(Sizes)
```

```python
#build volume array

Volumes = np.zeros([number,19])

for a in range(number):
    Volumes[a][0] = table_2['volume'][a]
```

```python
#add in volumes for Volume array for cooresponding voids
#each horizontal row is the same void

epsilon = 5

for a in range(number):
    for b in range(1,19):
        c = b + 2
        table = Table.read(file_name+str(c)+'.txt', format='ascii.commented_header')
        #print(table)
        for d in range(len(table['x'])):
            diffx = table_2['x'][a] - table['x'][d]
            diffy = table_2['y'][a] - table['y'][d]
            diffz = table_2['z'][a] - table['z'][d]
            diff = np.sqrt(diffx**2 + diffy**2 + diffz**2)
            if diff < epsilon:
                Volumes[a][b] = table['volume'][d]
    print('done with void ', a)
    print(Volumes[a])
```

```python
volumes = np.zeros([number,19])

for a in range(number):
    scale = Volumes[a][3]
    for b in range(19):
        volumes[a][b] = abs(Volumes[a][b]-scale) / scale
        if volumes[a][b]==1.0:
            volumes[a][b]='nan'
    print('done with void ', a)
    print(volumes[a])
```

```python
#Scale Volume values

for a in range(number):
    scale = Volumes[a][3]
    for b in range(19):
        Volumes[a][b] = Volumes[a][b] / scale
        if Volumes[a][b]==0.0:
            Volumes[a][b]='nan'
    print('done with void ', a)
    if Volumes[a][0]!=1.0:
        print(Volumes[a][0], Volumes[a][1], Volumes[a][3])
    elif Volumes[a][1]!=1.0:
        print(Volumes[a][0], Volumes[a][1], Volumes[a][3])
```

```python
Medians = np.zeros(19)
for a in range(19):
    temp = Volumes[:,a]
    temp = np.sort(temp)
    maximum = max(temp)
    for b in range(len(temp)):
        if temp[b]==maximum:
            m = b + 1
    size = len(temp) - m
    index = np.zeros(size, dtype=int)
    for b in range(len(index)):
        index[b] = m + b
    temp = np.delete(temp, index)
    Medians[a] = np.median(temp)
print(Medians)
```

```python
#plot each scaled volumes

%matplotlib notebook
plt.figure()
plt.plot(Sizes, Medians)
plt.xlabel('Grid Size')
plt.ylabel('Scaled Volumes')
plt.title('Volumes of Each Void and the Median for each Grid Size')

for a in range(number):
    volume = Volumes[a]
    size = Sizes
    ind = np.zeros(0, dtype=int)
    for b in range(19):
        if volume[b] == 0:
            ind = np.append(ind,size[b])
    size = np.delete(size, ind)
    volume = np.delete(volume, ind)
    #print(volume)
    plt.plot(size, volume, '#D9CBEB')
    #print('done with void ', a)

plt.plot(Sizes, Medians, 'blue')
plt.tight_layout();
```
