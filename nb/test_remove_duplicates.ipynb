```python
from astropy.table import Table

import numpy as np

import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
```

```python
spheres = Table.read('../data/vollim_dr7_cbp_102709_holes.txt', format='ascii.commented_header')
unique_spheres = Table.read('../data/vollim_dr7_cbp_102709_holes_unique.txt', format='ascii.commented_header')
```

```python
fig, ax = plt.subplots(figsize=(16,16))

patches = []

'''
for i in [0,1,2,3,4]:
    s = Circle((spheres['x'][i], spheres['y'][i]), spheres['radius'][i])
    patches.append(s)
'''
void = 1
holes = unique_spheres[unique_spheres['flag'] == void]
#holes.pprint()
for i in range(len(holes)):
    s = Circle((holes['y'][i], holes['z'][i]), holes['radius'][i])
    patches.append(s)

colors = 200*np.random.rand(len(patches))
p = PatchCollection(patches, alpha=0.2)
p.set_array(np.array(colors))
ax.add_collection(p)
fig.colorbar(p, ax=ax)

ax.axis('equal')

plt.show()
```
