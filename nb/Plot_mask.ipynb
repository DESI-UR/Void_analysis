Plot the survey mask generated by `VAST.VoidFinder`

```python
import pickle

import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook
```

# Mask file

```python
directory = '/Users/kellydouglass/Documents/Research/Voids/void_catalogs/SDSS/python_implementation/'

NSAmask_filename = 'NSA_main_mask.pickle'

temp_infile = open(directory + NSAmask_filename, 'rb')
NSA_mask, NSA_mask_resolution = pickle.load(temp_infile)
temp_infile.close()
```

```python
KIASmask_filename = 'kias1033_5_main_mask.pickle'

temp_infile = open(directory + KIASmask_filename, 'rb')
KIAS_mask, KIAS_mask_resolution = pickle.load(temp_infile)
temp_infile.close()
```

# Plot mask

```python
plt.figure(figsize=(12,12))

plt.imshow(NSA_mask.T, origin='lower')

plt.xlabel('ra [deg]')
plt.ylabel('dec [deg]')

plt.title('NSA v1.0.1')

plt.tight_layout();
```

```python
plt.figure(figsize=(12,12))

plt.imshow(KIAS_mask.T, origin='lower')

#ax = plt.gca()
#ax.set_xticklabels(np.arange(0,360,1./KIAS_mask_resolution))
#ax.set_yticklabels(np.arange(0,180,1./KIAS_mask_resolution))

plt.xlabel('ra [deg]')
plt.ylabel('dec [deg]')

plt.title('KIAS-VAGC')

plt.tight_layout();
```

```python
plt.figure(figsize=(12,12))

plt.imshow(NSA_mask.T != KIAS_mask.T, origin='lower')

plt.xlabel('ra [deg]')
plt.ylabel('dec [deg]')

plt.title('NSA - KIAS')

plt.tight_layout();
```
